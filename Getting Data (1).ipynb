# **Step-2**

```python
def stockPrice(company):
    ticker = company+".NS"
    stock = yf.Ticker(ticker)

    historical_data = stock.history(start= "2022-01-10",end = "2025-03-30", interval="1d")  # Data for the past month
    historical_data.to_csv(f"stock_data/raw/{company}.csv")
```

```python
for company in df["Symbol"]:
    stockPrice(company)
    print(company)
```
